**Key Concept:**
- Document vs Node

- In LlamaIndex:
  - Document = raw data
  - Node = chunked, processed unit

___
___

Raw Data → Documents → Nodes → Metadata → Ready for Indexing

#### Core Objects You MUST Understand

____

### Document

this is your raw container

```python
from llama_index.core import Document

doc = Document(
    text="This is some text",
    metadata={"source": "file1.txt"}
)

- A Document = one logical unit
- Not necessarily one file

___

### Node

- Node = small chunk of a document
- After parsing:
  - Document → Node1, Node2, Node3...

- Each node contains:
  - text chunk
  - metadata
  - relationships

___

### Types of Data You Can Ingest

___

### 1.File-Based Data
   - PDF
   - TXT
   - CSV
   - Markdown

**Loader:**

```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader("./data").load_data()

#### What happens internally:
- Reads files
- Converts → Documents

___

### 2.API Data
- News API
- Weather API
- Stock API

```python
import requests
from llama_index.core import Document

res = requests.get("https://api.example.com/data")
data = res.json()

documents = [Document(text=str(data))]

___

### 3.Database Data

```python
import sqlite3
from llama_index.core import Document

conn = sqlite3.connect("example.db")
cursor = conn.cursor()

rows = cursor.execute("SELECT * FROM logs").fetchall()

documents = [Document(text=str(row)) for row in rows]

- Convert rows → meaningful sentences

___

### 4.Streaming Data

Kafka → Consumer → Document → Index (incremental)

___

### 5.Web Scraping

```python
from llama_index.readers.web import SimpleWebPageReader

documents = SimpleWebPageReader().load_data(
    ["https://example.com"]
)

___
___
___

## Metadata

- This is where power comes from.

```python
Document(
    text="Blood sugar is high",
    metadata={
        "date": "2026-03-01",
        "type": "lab_report",
        "priority": "high"
    }
)

- Later you can:
  - filter
  - rank
  - route queries

___

## Custom Ingestion Pipeline

```python
from llama_index.core import Document

def custom_loader():
    docs = []

    # Example: multiple sources
    docs.append(Document(text="PDF content", metadata={"source": "pdf"}))
    docs.append(Document(text="API content", metadata={"source": "api"}))

    return docs

___

## Advanced Ingestion Concepts

- Data Cleaning
  - remove noise
  - normalize text
- Data Structuring
  - tables → sentences
  - JSON → readable format
- Multi-Modal (optional)
  - images (OCR)
  - audio → text

___

### Real Pipeline Example

```python
from llama_index.core import Document
import requests

documents = []

# PDF
documents.append(Document(
    text="Research paper content...",
    metadata={"source": "pdf"}
))

# API
news = requests.get("https://api.example.com/news").json()
documents.append(Document(
    text=str(news),
    metadata={"source": "api"}
))

# DB
documents.append(Document(
    text="Log: error in pipeline",
    metadata={"source": "db"}
))

___
