### Why Node Parsing Exists (Real Reason)

- LLMs cannot handle large raw documents effectively.

So we:
- break data into small meaningful chunks
- preserve context + structure
- That chunk = Node
  

#### Node = NOT just a chunk

- A Node contains:
   - text
   - metadata
   - relationships (optional)
   - position in document

#### Relationships Between Nodes

- Nodes can be linked:
   - Node A → Node B → Node C
- Types
   - previous / next
   - parent / child
- Used in:
   - tree index
   - graph index

```python
from llama_index.core.node_parser import SentenceSplitter

parser = SentenceSplitter(
    chunk_size=300,
    chunk_overlap=50
)

nodes = parser.get_nodes_from_documents(documents)

print(nodes[0].text)
print(nodes[0].metadata)


___

### Relationship types in LlamaIndex

- There are two categories of relationships:

- Automatically created (inbuilt)
- Explicitly created

___

#### Inbuilt relationships (default)

```python 
SentenceSplitter().get_nodes_from_documents(documents)
```

#### One-line clarity

- Sequential relationships are default; hierarchical relationships are intentional design.

- Final intuition
