### Common Mistakes

- Mistake 1: Not persisting
  - rebuilding index every run
  - slow + expensive
- Mistake 2:Mixing storage paths
  - corrupt index
- Mistake 3: Ignoring metadata storage
  - retrieval becomes weak
- Mistake 4: Thinking vector DB = full storage
  - WRONG:
  - You still need:
    - docstore
    - index structure

### What “Persistence” Means in LlamaIndex

- Persistence simply means:
```raw
Saving data so it survives after your program stops
```
- Example:
- Run app → build index → save to disk → close app  
Later → reopen app → load index
