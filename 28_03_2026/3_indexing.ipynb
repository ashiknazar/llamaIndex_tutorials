### Indexing (CORE OF LLAMAINDEX)

- Types of Indexes:
   - Vector Index (most common)
   - Tree Index
   - Keyword Index
   - Graph Index (important advanced topic)

- You must understand:
   - How nodes are stored
   - How retrieval works per index type

___

### What is Indexing (in LlamaIndex)?

- Indexing = converting Nodes into a searchable structure
- After this step:
   - your data becomes queryable
   - retrieval becomes fast + relevant
   

- Database index(fast lookup)
- but for semantic + structured data

___

- After parsing, you have:
```raw
Node1, Node2, Node3, ... Node1000
```

- LLM cannot read all nodes
- Linear search is slow
- No relevance ranking

___

#### What an Index Actually Does

- An index
  - Stores nodes
  - Organizes them
  - Enables fast retrieval


___
Full pipeline (important mental model)
```raw
Raw Data → Documents → Nodes → Embeddings → Index → Query Engine
```

___

### Step 1: Nodes → Embeddings

- Each Node is converted into a vector representation

- Node text:
"Machine learning is a subset of AI"

- Embedding:
[0.23, -0.91, 0.44, ...]

____

### Step 2: Storage inside Index

**It stores:**

- Node text
- Metadata
- Embeddings
- Relationships

___

### Types of Indexes

#### 1.VectorStoreIndex

- Stores embeddings in a vector database
- During query:
  - converts query → embedding
  - finds similar nodes

#### ListIndex (SimpleIndex)

- Stores nodes in sequence
- Reads one by one
- Rarely used in production

### 3. TreeIndex
- Organizes nodes in tree structure
- Summarizes at different levels

- Used for:
  - long documents
  - summarization tasks

### 4.Graph Index
- Nodes are connected
```raw
Node A → relates to → Node B
```


### 4. KeywordTableIndex
- Based on keyword matching
- No embeddings
- Old-school retrieval
